## 11.Выведите TOP-5 игроков, имеющих максимальный рост

```python
def height_to_inches(height):
    if pd.isna(height):
        return np.nan
    feet, inches = map(int, height.split('-'))
    return feet * 12 + inches

df['Height_inches'] = df['Height'].apply(height_to_inches)

top_5_tallest = df.nlargest(5, 'Height_inches')
display(top_5_tallest[['Name', 'Team', 'Height', 'Height_inches']])
```

## 14.Игроки какой команды образованны более других?

```python
df_copy=df.copy()
team_edu = (df_copy.groupby('Team').agg(
    total_players=('Name','count'),
    with_college=('College','count'),)
)
team_edu['college_rate']=team_edu['with_college']/team_edu['total_players']
team_edu=team_edu.sort_values('college_rate',ascending=False).head(1)
print('Самая образованная команда')
display(team_edu)
```

## 15.Есть ли связь между образованием и заработной платой игрока?

```python
df_copy=df.copy()
df_copy['has_college']=df_copy['College'].notna()
res=df_copy.groupby('has_college')['Salary'].agg(
    count='count',
    mean='mean',
    median='median'
)
display(res)
```
